```python
import numpy as np

# Concatenate in 1-D array

a = np.random.randint(2, 15, 6)
print(a)

b = np.random.randint(2, 15, 6)
print(b)

print(np.concatenate((a, b)))
```

```python
import numpy as np

# Concatenate in 2-D array

a = np.random.randint(2, 15, (2, 3))
print(a)

b = np.random.randint(2, 15, (2, 3))
print(b)

print(np.concatenate((a, b), axis=0))  # Row-wise

print(np.concatenate((a, b), axis=1))  # Column-wise
```
