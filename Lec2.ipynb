```python
lst = [1,2,3,'hello','c',[2,3]]
print(lst)
#append()
#to add element at the last
lst.append('hello')
print(lst)


#extend()
#to add a list
lst.extend('hello')
print(lst)

#insert()
#add element in between in a list
lst.insert(2,"value")
print(lst)
```

```python
#removing function from the list
#pop
#remove element from the last
lst = [1,2,3,'hello','c',[2,3]]
lst.pop()
print(lst)


#remove
#it remove element according to our need(basically remove the value of the index)
lst = [1,2,3,'hello','c',[2,3]]
lst.remove(2)
print(lst)

```

```python
#index
#it return the index
lst = [1,2,3,5,4,3,4,5,5,5]
print(lst.index(3))


#count
#count the element given in array acc to the given value
print(lst.count(5))
```

```python
#sort
lst = [1,2,3,5,4,3,4,5,5,5]
lst.sort()
#sort in same array
#we wnat to sort the list that why we print only lst
print(lst)




```

```python
#sum,min,max,sorted
lst = [1,2,3,5,4,3,4,5,8,5,5]
maximum=max(lst)
print(maximum)

minimum=min(lst)
print(minimum)

addition=sum(lst)
print(addition)

#it give new array after sorted
#after printing the lst it print the unsorted array
sequence=sorted(lst)
print(sequence)
print(lst)
```

```python
# tuple
# a object in python that not change a data
tp=(1,5,6,2,3,4)
a=tuple(sorted(tp))
print(a)
```
